```python

# Identifing numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns
```

```python
# Function to detect outliers using IQR
def detect_outliers(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    return outliers
```

```python
#checking for outliers in TAX column
outliers_tax = detect_outliers(df, 'TAX')
print("Number of outliers in TAX:", outliers_tax.shape[0])
```

Finding: There are no significant outliers present

```python
# Function to cap outliers
def cap_outliers(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[col] = np.where(data[col] < lower_bound, lower_bound,
                np.where(data[col] > upper_bound, upper_bound, data[col]))
    return data
```

```python
# Applying capping to all numeric columns
for col in numeric_cols:
    df = cap_outliers(df, col)

print("Outlier treatment complete.")
```

## Data preparation for association rule mining/ Build Transactions

This involves transforming the data into a transaction format suitable for association rule mining, focusing on fertilizers and seasons.

This is achieved by selecting the relevant columns and grouping rows so that each transaction = set of fertilisers applied to one plot in one season.


```python
# Combining location, block, and rep into one unique plot identifier
df['plot_id'] = df['locn'].astype(str) + "_" + df['block'].astype(str) + "_" + df['rep'].astype(str)
```

```python

# Grouping fertilisers into sets per plot per season
transactions = df.groupby(['plot_id', 'Sesn'])['ferT'].apply(list).reset_index()
```

```python
# Displaying the created transactions
print("Transactions (Fertilizers used per Season):")
display(transactions.head())
```

Now each transaction is a basket of fertilisers.

# Transaction Matrix
This involves converting the list of fertilizers for each season into a one-hot encoded format for association rule mining.

Apriori algorithm requires binary form (1 if fertiliser is present, 0 if not)


```python
from mlxtend.preprocessing import TransactionEncoder

# Converting transactions to list of lists
transaction_list = transactions['ferT'].tolist()
```

```python
# encode
te = TransactionEncoder()
te_array = te.fit(transaction_list).transform(transaction_list)
df_encoded = pd.DataFrame(te_array, columns=te.columns_)

# Displaying the encoded data
print("\n Encoded Transactions:")
df_encoded.head()
```

Now each row shows which fertilisers appear in that plot-season.

## Apply association rule mining using Apriori Algorithm

this involves extracting frequent itemsets (combinations of fertilisers) that appear often.


```python
from mlxtend.frequent_patterns import apriori

# Minimum support threshold = 0.5 (appear in at least 50% of transactions)
# Finding frequent itemsets
frequent_itemsets = apriori(df_encoded, min_support=0.5, use_colnames=True)

frequent_itemsets.sort_values(by="support", ascending=False, inplace=True)
frequent_itemsets.head()
```

All fertilisers co-occur in every transaction → perfect association.

Generate Association Rules

By deriving rules of the form A → B with metrics support, confidence, lift.

```python
from mlxtend.frequent_patterns import association_rules

# Generating association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
rules[['antecedents','consequents','support','confidence','lift']].head()
```

Presence of any fertiliser perfectly predicts the others.
